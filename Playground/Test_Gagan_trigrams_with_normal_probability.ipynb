```python
import pandas as pd
import re
import math
import operator
from collections import defaultdict
from decimal import Decimal
```

```python
df = pd.read_csv('../OriginalDataSet/training-tweets.txt', encoding='utf-8', error_bad_lines=False, sep='\t', nrows=18000)
df.columns = ['TweetID', 'UserID', 'Language', "Tweet"]
```

```python
df.head(50)
df.shape
```

```python
_df0 = df[['Language', 'Tweet']].copy()
pattern = re.compile('[a-z]')
_df0.shape
# _df0.head()
```

```python
_df0.head()
```

```python
trainDict = defaultdict(list)
for index, row in _df0.iterrows():
    sentence = ''
    tweet = row['Tweet']
    tweet = tweet.lower()
    for letter in tweet:
        if pattern.match(letter):
            sentence = sentence + letter
    trainDict[row['Language']].append(sentence)
```

```python
print(trainDict.keys())
```

```python
enDict = {}
esDict = {}
euDict = {}
caDict = {}
ptDict = {}
glDict = {}
delta = 0.5
for key, value in trainDict.items():
    #get the dictionary corresponding to the language(key)
    trigramCount = 0
    if 'en' in key:
        for eachTweet in value:
            for i in range(len(eachTweet)-2):
                first = eachTweet[i]
                second = eachTweet[i+1]
                third = eachTweet[i+2]
                if ((not first.isspace()) and (not second.isspace()) and (not third.isspace()) ):
                    trigram = first + second + third
                    if trigram in enDict:
                        
#                         enDictLevel1 = enDictLevel0[first]
#                         enDictLevel2 = enDictLevel1[second]
#                         count = enDictLevel2[third]                         
                        
                        count = enDict[trigram]
                        count += 1
                        enDict[trigram] = count
                        trigramCount += 1
                    else:
                        enDict[trigram] = 1 + delta
                        trigramCount += 1
        enDict['NON'] = delta
        enDict = {k: format((v/trigramCount), '.15f') for k, v in enDict.items()}
    elif 'es' in key:
        for eachTweet in value:
            for i in range(len(eachTweet)-2):
                first = eachTweet[i]
                second = eachTweet[i+1]
                third = eachTweet[i+2]
                if ((not first.isspace()) and (not second.isspace()) and (not third.isspace()) ):
                    trigram = first + second + third
                    if trigram in esDict:
                        count = esDict[trigram]
                        count += 1
                        esDict[trigram] = count
                        trigramCount += 1
                    else:
                        esDict[trigram] = 1 + delta
                        trigramCount += 1
        esDict['NON'] = delta
        esDict = {k: format((v/trigramCount), '.15f') for k, v in esDict.items()}
    elif 'eu' in key:
        for eachTweet in value:
            for i in range(len(eachTweet)-2):
                first = eachTweet[i]
                second = eachTweet[i+1]
                third = eachTweet[i+2]
                if ((not first.isspace()) and (not second.isspace()) and (not third.isspace()) ):
                    trigram = first + second + third
                    if trigram in euDict:
                        count = euDict[trigram]
                        count += 1
                        euDict[trigram] = count
                        trigramCount += 1
                    else:
                        euDict[trigram] = 1 + delta
                        trigramCount += 1
        euDict['NON'] = delta
        euDict = {k: format((v/trigramCount), '.15f') for k, v in euDict.items()}
    elif 'ca' in key:
        for eachTweet in value:
            for i in range(len(eachTweet)-2):
                first = eachTweet[i]
                second = eachTweet[i+1]
                third = eachTweet[i+2]
                if ((not first.isspace()) and (not second.isspace()) and (not third.isspace()) ):
                    trigram = first + second + third
                    if trigram in caDict:
                        count = caDict[trigram]
                        count += 1
                        caDict[trigram] = count
                        trigramCount += 1
                    else:
                        caDict[trigram] = 1 + delta
                        trigramCount += 1
        caDict['NON'] = delta
        caDict = {k: format((v/trigramCount), '.15f') for k, v in caDict.items()}
    elif 'pt' in key:
        for eachTweet in value:
            for i in range(len(eachTweet)-2):
                first = eachTweet[i]
                second = eachTweet[i+1]
                third = eachTweet[i+2]
                if ((not first.isspace()) and (not second.isspace()) and (not third.isspace()) ):
                    trigram = first + second + third
                    if trigram in ptDict:
                        count = ptDict[trigram]
                        count += 1
                        ptDict[trigram] = count
                        trigramCount += 1
                    else:
                        ptDict[trigram] = 1 + delta
                        trigramCount += 1
        ptDict['NON'] = delta
        ptDict = {k: format((v/trigramCount), '.15f') for k, v in ptDict.items()}
    elif 'gl' in key:
        for eachTweet in value:
            for i in range(len(eachTweet)-2):
                first = eachTweet[i]
                second = eachTweet[i+1]
                third = eachTweet[i+2]
                if ((not first.isspace()) and (not second.isspace()) and (not third.isspace()) ):
                    trigram = first + second + third
                    if trigram in glDict:
                        count = glDict[trigram]
                        count += 1
                        glDict[trigram] = count
                        trigramCount += 1
                    else:
                        glDict[trigram] = 1 + delta
                        trigramCount += 1
        glDict['NON'] = delta
        glDict = {k: format((v/trigramCount), '.15f') for k, v in glDict.items()}
```
