```python
model = {'en':enDict, 'es':esDict, 'eu':euDict, 'ca':caDict, 'pt':ptDict, 'gl':glDict}
df = pd.read_csv('../OriginalDataSet/training-tweets.txt', encoding='utf-8', error_bad_lines=False, sep='\t', nrows=50)
df.columns = ['TweetID', 'UserID', 'Language', "Tweet"]
_df0 = df[['TweetID', 'Language', 'Tweet']].copy()
# pattern = re.compile('[a-z ]')
# _df0.head()
```

```python
for index, row in _df0.iterrows():
    enProb = 0
    euProb = 0
    esProb = 0
    caProb = 0
    ptProb = 0
    glProb = 0
    tweetID = row['TweetID']
    lang = row['Language']
    tweet = row['Tweet']
    for i in range(len(tweet)-1):
        first = tweet[i]
        second = tweet[i+1]
        bigram = first + second
        
        if bigram in enDict:
            enProb = enDict[bigram] + enProb
        elif pattern.match(bigram):
            enProb = enDict['NON'] + enProb
        else:
            pass
        
        if bigram in euDict:
            euProb = euDict[bigram] + euProb
        elif pattern.match(bigram):
            euProb = euDict['NON'] + euProb
        else:
            pass
            
        if bigram in esDict:
            esProb = esDict[bigram] + esProb
        elif pattern.match(bigram):
            esProb = esDict['NON'] + esProb
        else:
            pass

        if bigram in caDict:
            caProb = caDict[bigram] + caProb
        elif pattern.match(bigram):
            caProb = caDict['NON'] + caProb
        else:
            pass
            
        if bigram in ptDict:
            ptProb = ptDict[bigram] + ptProb
        elif pattern.match(bigram):
            ptProb = ptDict['NON'] + ptProb
        else:
            pass
        
        if bigram in glDict:
            glProb = glDict[bigram] + glProb
        elif pattern.match(bigram):
            glProb = glDict['NON'] + glProb
        else:
            pass
    probability = {'en':enProb, 'eu':euProb, 'es':esProb, 'ca':caProb, 'pt':ptProb, 'gl':glProb}
    result = max(probability, key=probability.get)
#     print('tweet: ', tweet,'\nenProb:', enProb, '\neuProb:', euProb, '\nesProb:', esProb, '\ncaProb:', caProb
#          ,'\nptProb:', ptProb, '\nglProb:', glProb)
#     print('probable language', result)
    print(tweetID, '  ', result, '  ', probability[result], '  ', lang)
```
